4.  What is the difference between abstraction and encapsulation?
- Difference Between Abstraction and Encapsulation
1. Definition:

Abstraction: Hides implementation details and only shows the essential features.

Encapsulation: Wraps data and methods into a single unit and restricts direct access.

2. Purpose:

Abstraction: Reduces complexity by showing only relevant details.

Encapsulation: Protects data from unauthorized access and modification.

3. Implementation:

Abstraction: Achieved using abstract classes and interfaces.

Encapsulation: Achieved using access modifiers (private, protected, public).

4. Focus:

Abstraction: Focuses on what an object does.

Encapsulation: Focuses on how data is controlled.

5. Real-World Example:

Abstraction: A car's steering wheel allows driving without knowing how the engine works.

Encapsulation: A car's engine is hidden, preventing direct modification.


```python
#example of Abstraction
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car is starting with a key!")

car = Car()
car.start()
```

```python
#example of encapsulation
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
account.deposit(500)
print(account.get_balance())

```

5.  What are dunder methods in Python?
- Dunder (Double Underscore) Methods in Python
Dunder methods (short for "double underscore methods") are special methods in Python that have double underscores (__) at the beginning and end of their names. They are also called magic methods because they enable built-in functionalities like object creation, string representation, arithmetic operations, and more.

```python
#examples of Dunder methods
#1. __init__
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Ashif", 21)
print(p1.name)
```

```python
#2. __str__ and __repr__
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} is {self.age} years old"

    def __repr__(self):
        return f"Person('{self.name}', {self.age})"

p1 = Person("Ashif", 21)
print(str(p1))
print(repr(p1))
```

```python
#3. __len__
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __len__(self):
        return self.pages

b1 = Book("Python OOPS", 350)
print(len(b1))
```

6.  Explain the concept of inheritance in OOP?
- Inheritance in Object-Oriented Programming (OOP)
Inheritance is a fundamental concept in OOP that allows a class (child/derived class) to inherit properties and methods from another class (parent/base class). It promotes code reusability and hierarchical relationships between classes.

Types of Inheritance in Python

(1) Single Inheritance â€“ One class inherits from another.

(2) Multiple Inheritance â€“ A class inherits from multiple parent classes.

(3) Multilevel Inheritance â€“ A child class inherits from another child class.

(4) Hierarchical Inheritance â€“ Multiple child classes inherit from a single parent.

(5) Hybrid Inheritance â€“ A combination of multiple types of inheritance.



```python
#examples
#Single Inheritence
class Animal:
    def speak(self):
        return "I make a sound"

class Dog(Animal):
    def speak(self):
        return "Woof!"

d = Dog()
print(d.speak())
```

```python
#Multiple Inheritence
class Father:
    def trait(self):
        return "Tall"

class Mother:
    def trait(self):
        return "Brown Eyes"

class Child(Father, Mother):
    pass

c = Child()
print(c.trait())
```

```python
#Multilevel Inheritance
class Grandparent:
    def family_name(self):
        return "Sarkar"

class Parent(Grandparent):
    pass

class Child(Parent):
    pass

c = Child()
print(c.family_name())
```

```python
#Hierarchical Inheritance
class Vehicle:
    def mode(self):
        return "Transport"

class Car(Vehicle):
    def wheels(self):
        return 4

class Bike(Vehicle):
    def wheels(self):
        return 2

c = Car()
b = Bike()
print(c.mode(), c.wheels())
print(b.mode(), b.wheels())
```

```python
#Hybrid Inheritance
class A:
    def method_A(self):
        return "A"

class B(A):
    def method_B(self):
        return "B"

class C(A):
    def method_C(self):
        return "C"

class D(B, C):
    def method_D(self):
        return "D"

d = D()
print(d.method_A(), d.method_B(), d.method_C(), d.method_D())

```

7. What is polymorphism in OOP?
- Polymorphism in Object-Oriented Programming (OOP)
Polymorphism means "many forms" and allows objects of different classes to be treated as objects of a common superclass. It enables one interface to be used for different underlying data types.

Types of Polymorphism in Python

(1) Method Overriding (Runtime Polymorphism)

(2) Method Overloading (Compile-time Polymorphism - Not directly supported in Python)

(3) Operator Overloading

(4) Polymorphism with Functions and Classes

```python
#Example
#Method Overriding (Runtime Polymorphism)
class Animal:
    def speak(self):
        return "I make a sound"

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

animals = [Dog(), Cat()]
for animal in animals:
    print(animal.speak())
```

```python
#Method Overloading (Not Directly Supported in Python)
class Math:
    def add(self, a, b, c=0):
        return a + b + c

m = Math()
print(m.add(2, 3))
print(m.add(2, 3, 4))
```

```python
#Operator Overloading
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def __add__(self, other):
        return BankAccount(self.balance + other.balance)

a1 = BankAccount(1000)
a2 = BankAccount(2000)

a3 = a1 + a2
print(a3.balance)
```

```python
# Polymorphism with Functions and Classes
class Car:
    def fuel_type(self):
        return "Petrol"

class ElectricCar:
    def fuel_type(self):
        return "Electric"

def show_fuel(vehicle):
    print(vehicle.fuel_type())

c1 = Car()
c2 = ElectricCar()

show_fuel(c1)
show_fuel(c2)
```

10. What are class and static methods in Python?
- Class Methods vs. Static Methods in Python
Python provides two special types of methods in addition to instance methods:

Class Methods (@classmethod)

Static Methods (@staticmethod)

Both are used when we don't need to work with instance-specific data.

(1) Class Method (@classmethod)
A class method operates on the class itself, not on an instance. It can modify class-level variables and is defined using @classmethod.

```python
#Example
class Employee:
    company = "Google"

    @classmethod
    def change_company(cls, new_name):
        cls.company = new_name

print(Employee.company)
Employee.change_company("Microsoft")
print(Employee.company)
```

(2) Static Method (@staticmethod)
A static method does not operate on the instance or the class. It behaves like a regular function inside a class and is defined using @staticmethod.

```python
#example
class MathOperations:
    @staticmethod
    def add(x, y):
        return x + y

print(MathOperations.add(5, 10))
```

11. What is method overloading in Python?
- Method Overloading in Python
ðŸ”¹ What is Method Overloading?
Method Overloading allows multiple methods with the same name but different parameters (different number or type of arguments).

ðŸ”¹ Python does NOT support true method overloading like Java or C++.


```python
# Simulating Method Overloading Using Default Arguments
class Math:
    def add(self, a, b=0, c=0):
        return a + b + c

m = Math()
print(m.add(2))
print(m.add(2, 3))
print(m.add(2, 3, 4))
```

```python
# Using args for Dynamic Arguments
class Math:
    def add(self, *args):
        return sum(args)

m = Math()
print(m.add(2, 3))
print(m.add(2, 3, 4, 5))
print(m.add(10, 20, 30))
```

```python
#Using @singledispatch for Type-Based Overloading
from functools import singledispatch

@singledispatch
def display(value):
    print("Default:", value)

@display.register(int)
def _(value):
    print("Integer:", value)

@display.register(str)
def _(value):
    print("String:", value)

display(10)
display("Hello")
display(3.14)
```

13.  What is a property decorator in Python?
- The @property decorator in Python is used to define getter methods that allow access to private attributes like a normal attribute, without needing explicit getter and setter methods.

 It helps in encapsulation by controlling how attributes are accessed and modified.

 Instead of using obj.get_value(), we can use obj.value like a normal attribute.



```python
#Basic example
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

p = Person("Ashif")
print(p.name)
```

```python
#Using @property with Setter
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, new_name):
        if len(new_name) < 3:
            raise ValueError("Name must be at least 3 characters!")
        self._name = new_name

p = Person("Ashif")
print(p.name)

p.name = "Zufi"
print(p.name)

```

```python
#Using @property with Deleter
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @name.deleter
    def name(self):
        print("Deleting name...")
        del self._name

p = Person("Ashif")
del p.name
```

14. Why is polymorphism important in OOP?
- Polymorphism is one of the four pillars of Object-Oriented Programming (OOP) (along with Encapsulation, Abstraction, and Inheritance). It allows a single interface to be used for different types of objects, making the code more flexible, reusable, and scalable.

(1) Increases Code Reusability and Maintainability

Instead of writing separate functions/methods for each data type or class, polymorphism allows one function to work with multiple object types.

```python
#Example
class Dog:
    def sound(self):
        return "Woof!"

class Cat:
    def sound(self):
        return "Meow!"

def make_sound(animal):
    return animal.sound()

d = Dog()
c = Cat()

print(make_sound(d))
print(make_sound(c))
```

Polymorphism allows methods to be called on objects without knowing their exact class at compile time.

```python
#Example
class Parent:
    def show(self):
        return "I am the parent"

class Child(Parent):
    def show(self):
        return "I am the child"

def display(obj):
    return obj.show()

p = Parent()
c = Child()

print(display(p))
print(display(c))
```

(3) Simplifies Code Using Operator Overloading

Polymorphism allows operators like +, *, == to work differently for different data types.

```python
#Example
print(5 + 10)
print("Hello" + " World")
print([1, 2] + [3, 4])
```

Polymorphism allows new classes to be added without modifying existing code.

```python
#Example
class Bird:
    def sound(self):
        return "Chirp"
b = Bird()
print(make_sound(b))
```

15.  What is an abstract class in Python?
- An abstract class in Python is a class that cannot be instantiated directly and is meant to be a blueprint for other classes. It contains abstract methods, which must be implemented by any subclass.

Python provides the ABC (Abstract Base Class) module to define abstract classes.

How to Define an Abstract Class?
We use:

from abc import ABC, abstractmethod

Inheriting from ABC

Marking abstract methods with @abstractmethod

```python
#Example
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        return "Woof!"

class Cat(Animal):
    def sound(self):
        return "Meow!"

d = Dog()
print(d.sound())

c = Cat()
print(c.sound())
```

- Abstract Class with Concrete Methods

Abstract classes can also have normal (concrete) methods that subclasses inherit.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def fuel_type(self):
        return "Petrol or Diesel"

    @abstractmethod
    def max_speed(self):
        pass

class Car(Vehicle):
    def max_speed(self):
        return "200 km/h"

c = Car()
print(c.fuel_type())
print(c.max_speed())
```

- Abstract Class with Class Methods

We can use @abstractmethod with @classmethod, @staticmethod, or @property.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @classmethod
    @abstractmethod
    def shape_type(cls):
        pass

class Circle(Shape):
    @classmethod
    def shape_type(cls):
        return "This is a Circle"

print(Circle.shape_type())
```

24.  What is method chaining in Python OOP?
- Method chaining is a technique where you call multiple methods on the same object in a single line, one after the other. It works by having each method return self, so you can keep chaining more method calls.

```python
#syntax example
class Person:
    def __init__(self, name):
        self.name = name
        self.age = None
        self.city = None

    def set_age(self, age):
        self.age = age
        return self

    def set_city(self, city):
        self.city = city
        return self

    def show(self):
        print(f"{self.name}, Age: {self.age}, City: {self.city}")
        return self


p = Person("Ashif").set_age(21).set_city("Mumbai").show()
```

```python
#Eample
class QueryBuilder:
    def __init__(self):
        self.query = "SELECT * FROM table"

    def where(self, condition):
        self.query += f" WHERE {condition}"
        return self

    def order_by(self, field):
        self.query += f" ORDER BY {field}"
        return self

    def build(self):
        return self.query

q = QueryBuilder().where("age > 25").order_by("name").build()
print(q)
```

```python
#2.  Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
#from it and implement the area() method in both.
from abc import ABC, abstractmethod
import math

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

shapes = [
    Circle(5),
    Rectangle(4, 6)
]

for shape in shapes:
    print(f"Area: {shape.area():.2f}")
```

```python
#3.  Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
#and further derive a class ElectricCar that adds a battery attribute

  #Multi-Level Inheritance: Vehicle âž Car âž ElectricCar
  # Base class
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type

    def show_type(self):
        print(f"Vehicle Type: {self.vehicle_type}")

# Derived class from Vehicle
class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)
        self.brand = brand

    def show_brand(self):
        print(f"Car Brand: {self.brand}")

# Further derived class from Car
class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery_capacity):
        super().__init__(vehicle_type, brand)
        self.battery_capacity = battery_capacity

    def show_battery(self):
        print(f"Battery Capacity: {self.battery_capacity} kWh")

# Testing
tesla = ElectricCar("Four-wheeler", "BMW", "84")

tesla.show_type()
tesla.show_brand()
tesla.show_battery()
```

```python
#4.  Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes
#Sparrow and Penguin that override the fly() method.

# Base class
class Bird:
    def fly(self):
        print("Some bird is flying...")

# Derived class 1
class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high in the sky!")

# Derived class 2
class Penguin(Bird):
    def fly(self):
        print("Penguins can't fly, but they swim like pros!")

# Function to demonstrate polymorphism
def make_it_fly(bird):
    bird.fly()

# Testing
birds = [Sparrow(), Penguin()]

for b in birds:
    make_it_fly(b)
```

```python
#5.  Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
#balance and methods to deposit, withdraw, and check balance.

#BankAccount
class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited â‚¹{amount}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew â‚¹{amount}")
        else:
            print("Insufficient balance!")

    def check_balance(self):
        print(f"Current Balance: â‚¹{self.__balance}")

# Testing
account = BankAccount(1000)

account.deposit(500)
account.withdraw(300)
account.check_balance()


```

```python
#6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar
#and Piano that implement their own version of play()

# Base class
class Instrument:
    def play(self):
        print("Playing some instrument...")

# Derived class 1
class Guitar(Instrument):
    def play(self):
        print("Strumming the guitar ")

# Derived class 2
class Piano(Instrument):
    def play(self):
        print("Playing the piano ")

def start_music(instrument):
    instrument.play()

# Testing
instruments = [Guitar(), Piano()]

for i in instruments:
    start_music(i)
```

```python
#7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static
#method subtract_numbers() to subtract two numbers

class MathOperations:

    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

# Testing the methods
print("Addition:", MathOperations.add_numbers(10, 5))
print("Subtraction:", MathOperations.subtract_numbers(10, 5))
```

```python
#10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
#vectors.

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

# Testing
v1 = Vector(2, 5)
v2 = Vector(4, 3)

v3 = v1 + v2

print("Vector 1:", v1)
print("Vector 2:", v2)
print("Sum:", v3)
```

```python
#12.  Implement a class Student with attributes name and grades. Create a method average_grade() to compute
#the average of the grades
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)


# Example usage:
student1 = Student("Ashif", [85, 90, 78, 92])
print(f"{student1.name}'s average grade is: {student1.average_grade()}")

student2 = Student("Zufii", [])
print(f"{student2.name}'s average grade is: {student2.average_grade()}")
```

```python
#14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
#and hourly rate. Create a derived class Manager that adds a bonus to the salary
# Base class
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate


# Derived class
class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        base_salary = super().calculate_salary()
        return base_salary + self.bonus


# Example usage:
emp = Employee("Ashif", 40, 25)
print(f"{emp.name}'s salary: {emp.calculate_salary()}")

mgr = Manager("Zufii", 40, 30, 500)
print(f"{mgr.name}'s salary with bonus: {mgr.calculate_salary()}")
```

```python
#16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
#implement the sound() method
from abc import ABC, abstractmethod

# Abstract base class
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

# Derived class Cow
class Cow(Animal):
    def sound(self):
        return "Moo"

# Derived class Sheep
class Sheep(Animal):
    def sound(self):
        return "Baa"

# Example usage:
cow = Cow()
sheep = Sheep()

print(f"Cow sound: {cow.sound()}")
print(f"Sheep sound: {sheep.sound()}")
```

```python
#17.  Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that
#returns a formatted string with the book's details
class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"'{self.title}' by {self.author} (Published in {self.year_published})"

# Example usage:
book1 = Book("1984", "George Orwell", 1949)
print(book1.get_book_info())
```
